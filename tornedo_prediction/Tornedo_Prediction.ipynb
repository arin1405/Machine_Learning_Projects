```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('1950-2017_all_tornadoes.csv')
```

```python
df.head()
```

```python
df = df[df.yr > 2016]
```

```python
df['barometricPressure'] = 0
df['dewpoint'] = 0
df['elevation'] = 0
df['heatIndex'] = 0
df['relativeHumidity'] = 0
df['seaLevelPressure'] = 0
df['visibility'] = 0
df['windDirection'] = 0
df['windSpeed'] = 0
```

```python
from noaa_sdk import noaa
```

```python
n = noaa.NOAA()
osm = noaa.OSM()
```

```python
code_country = osm.get_postalcode_country_by_lan_lon(43.951,-88.0424)
```

```python
observations = n.get_observations('11365','us',start='2018-05-15',end='2018-05-30')
```

```python
next(observations)
```

```python
for item in n.get_observations('11365','us',start='2017-01-01',end='2018-05-15'):
    #print()
    barometricPressure = item['barometricPressure']['value']
    dewpoint = item['dewpoint']['value']
    elevation = item['elevation']['value']
    heatIndex = item['heatIndex']['value']
    relativeHumidity = item['relativeHumidity']['value']
    seaLevelPressure = item['seaLevelPressure']['value']
    visibility = item['visibility']['value']
    windDirection = item['windDirection']['value']
    windSpeed = item['windSpeed']['value']
    dt = item['timestamp'].split('T')[0].strip()
    
    df.loc[df['date'] == dt, 'barometricPressure'] = barometricPressure
    df.loc[df['date'] == dt, 'dewpoint'] = dewpoint
    df.loc[df['date'] == dt, 'elevation'] = elevation
    df.loc[df['date'] == dt, 'heatIndex'] = heatIndex
    df.loc[df['date'] == dt, 'relativeHumidity'] = relativeHumidity
    df.loc[df['date'] == dt, 'seaLevelPressure'] = seaLevelPressure
    df.loc[df['date'] == dt, 'visibility'] = visibility
    df.loc[df['date'] == dt, 'windDirection'] = windDirection
    df.loc[df['date'] == dt, 'windSpeed'] = windSpeed
    
```
